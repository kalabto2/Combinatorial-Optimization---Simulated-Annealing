Import modules

```python
import subprocess
import matplotlib.pyplot as plt
import os
from time import time
from statistics import mean
import numpy as np
```

Helping fuctions

```python
def timer(func):
    def wrap_func(*args, **kwargs):
        t1 = time()
        result = func(*args, **kwargs)
        t2 = time()
        print(f'Function {func.__name__!r} executed in {(t2-t1):.4f}s')
        return result
    return wrap_func
```

Calculate instance via C++ program

```python
@timer
def calculate(filepath: str, init_temperature: int=None, equilibrium: int=None, frozen_C: int=None, C: float=None, white_box=False):
    commands = ["python3", "simulated_annealing.py", filepath]
    if white_box:
        commands.append("-w")
    if init_temperature:
        commands += ["-t", str(init_temperature)]
    if equilibrium:
        commands += ["-e", str(equilibrium)]
    if frozen_C:
        commands += ["-f", str(frozen_C)]
    if C:
        commands += ["-c", str(C)]
    
    proc = subprocess.Popen(commands, stdout=subprocess.PIPE)
    output = proc.stdout.read().decode('UTF-8')
    return output
```

## White Box phase

Define functions

```python
def print_weight_per_iteration(y, optimal_weight: int, y_description="cost"):
    x = range(len(y))

    plt.plot(x, y, label="sim. aneal. states")
#     plt.plot(x, [optimal_weight] * len(y), label="optimal price")
    plt.xlabel("# of iterations")
    plt.ylabel(y_description)
    # plt.title(f"Test robustnosti pro {method.name}")  # add title
    plt.legend(bbox_to_anchor=(1,1), loc="upper left")
    plt.show()
```

```python
def load_optimal_result(filepath: str):
    f = open(filepath, "r")
    file_content = f.read()
    lines = file_content.split('\n')
    
    optimal_results = dict()
    for line in lines:
        instance_id = line.split(' ')[0]
        optimal_results[instance_id] = line
        
    return optimal_results     
```

Do white box phase

```python
instance_ids = ["uf20-02", "uf20-0200", "uf20-0206", "uf20-0199", "uf20-0113", "uf20-0114", "uf20-098",
               "uf20-0149", "uf20-0152", "uf20-0128"]
init_temperature = 15000
frozen = 100
equilibrium = 200
C = 0.9

sat_rate = []
weight_rate = []
for instance_id in instance_ids:
    output = calculate("sets/wuf-M1/wuf20-78-M1/w" + instance_id + ".mwcnf", 
                       init_temperature,
                       equilibrium,
                       frozen,
                       C,
                       white_box=True)
    [result, brief, iterations, clauses_sat, _] = output.split("\n")
    y = iterations.split(' ')
    y = list(map(lambda x: float(x), y))
    y2 = clauses_sat.split(' ')
    y2 = list(map(lambda x: float(x), y2))

    optimal = int(load_optimal_result("sets/wuf-M1/wuf20-78-M1-opt.dat")[instance_id].split(' ')[1])
    print(load_optimal_result("sets/wuf-M1/wuf20-78-M1-opt.dat")[instance_id])
    print(result)
    print(brief)
    print_weight_per_iteration(y, optimal, y_description="cost")
    print_weight_per_iteration(y2, optimal, y_description="clauses sat rate")
    
    sat_rate.append(int(brief.split(' ')[1]) / int(brief.split(' ')[2]))
    weight_rate.append(int(brief.split(' ')[0]) / optimal)
    
# ============ SUMMARY ============

print("\n\n====================================\n")
print("Avg. sat rate:", np.mean(sat_rate))
print("Avg. weight rate:", np.mean(weight_rate))
```

## Black box phase

Calculate sets

```python
def check_set(set_filepath: str,
             init_temperature: int,
             equilibrium: int,
             frozen_C: int,
             C: float
             ):
    for (root,dirs,files) in os.walk(set_filepath, topdown=True):
        print("Calculating", root, "...")
        if root == set_filepath:
            continue
        
        # calculate and save solutions and time
        results = []
        for file in files:
            output = calculate(root + "/" + file, init_temperature, equilibrium, frozen_C, C)
            [result, brief, _] = output.split('\n')
            results.append(result)
        
        # save results
        f = open(root + "-res.dat", "w")
        f.write("\n".join(results))
```

```python
check_set("sets/wuf-M1", 1000, 150, 10, 0.9)
```

Analyze results

```python
def analyze_set(set_filepath: str):
    f = open(set_filepath + "-res.dat", "r")
    content = f.read()
    results = content.split("\n")
    optimal_results = load_optimal_result(set_filepath + "-opt.dat")
    
    relative_error = []
    for result in results:
        inctance_id = result.split(" ")[0]
        optimal_weight = int(optimal_results[inctance_id].split(" ")[1])
        result_weight = int(result.split(" ")[1])
        relative_error.append(abs(optimal_weight - result_weight) / optimal_weight)
        
    mean_relative_error = mean(relative_error)
    max_relative_error = max(relative_error)
    var_relative_error = np.var(relative_error)
    
    print(mean_relative_error, max_relative_error, var_relative_error)
```

```python
analyze_set("wuf-M1/wuf20-78-M1")
```
